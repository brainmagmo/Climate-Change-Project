```python
# Dependencies and setup
import pandas as pd
import pandas as pd
import numpy as np
import os
```

```python
# Use os.join.path so any OS can run this code
file_path = os.path.join("..", "Resources", "population_density_by_country.csv")
world_pop = os.path.join("..", "Resources", "world population.csv")

# Read the csv into a data frame and display the first few rows
population_df = pd.read_csv(file_path, skiprows=4)
population_df.head()
```

```python
# We only want data for years 1986 - 2016 so we will remove any columns that aren't necessary 
population_df.drop(population_df.loc[:,"Indicator Name":"1985"], inplace=True, axis=1)

# Drop the remaining rows at the back end from 2017 to 2019
population_df.drop(population_df.loc[:,"2017":"Unnamed: 64"], inplace=True, axis=1)

# Display the first few rows
population_df.head()
```

```python
# Find all the rows with missing data
missing_data = population_df[population_df.isnull().any(axis=1)]

# Display the missing data
missing_data.head()
```

```python
# Drop the rows with missing data by the row index
population_df.drop(index=missing_data.index, inplace=True)

# Display the clean data
clean_data = population_df

# Save the clean data to a CSV
clean_data.to_csv("../Cleaned Data/population_density_clean.csv", index=False)

# Display the clean data
clean_data
```
