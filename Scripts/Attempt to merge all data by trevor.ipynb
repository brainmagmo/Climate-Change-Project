```python
data_dfs = []
for data_file in data_files:
    data_dfs.append(
        pd.read_csv(directory + data_file)
    )
    print(data_file)
    
for df in data_dfs:
    print(df.columns, '\n\n', df.size)
```

```python
def trim(string):
    return string.strip()
for index, row in data_dfs[0].iterrows():
    row['Location'] = row['Location'].strip()
    
data_dfs[0].Location = data_dfs[0].Location.map(trim)
data_dfs[0]
```

```python
consumption_countries = data_dfs[0].Location.unique()
consumption_countries
```

```python
density_countries = data_dfs[2]['Country Name'].unique()
density_countries
```

```python

def Diff(li1, li2): 
    '''
    # Python code t get difference of two lists 
    # Using set() 
    '''
    return (list(set(li1).symmetric_difference(set(li2))))
```

```python
delta = Diff(consumption_countries, density_countries)
delta
```

```python
'''
what questions do we want to answer?
'''
```

```python
import difflib
```

```python
possible_duplicates = []
for thing in delta:
    matches = difflib.get_close_matches(thing, delta)
    if len(matches) > 1:
        #print(matches)
        possible_duplicates.append(matches)
```

```python
pos_dup = []
for group in possible_duplicates:
    group.sort()
    
possible_duplicates.sort()
for group in possible_duplicates:
    if not (group in pos_dup):
        pos_dup.append(group)
pos_dup
```

```python
def slst_has_bad(lst, bad):
    '''
    finds if any string in a list contains and string in the list of strings bad
    '''
    for string in lst:
        for bad_string in bad:
            if string.find(bad_string) > -1:
                return True
        
    return False
    
dirty = ['&', 'income', 'dividend', 'only', 'total']
trash = [group for group in pos_dup if slst_has_bad(group, dirty)]
clean = [group for group in pos_dup if not slst_has_bad(group, dirty)]
            
trash
```
