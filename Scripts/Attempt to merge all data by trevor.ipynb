```python
directory = '../Cleaned Data/'
files = os.listdir(directory)
```

```python
data_files = []
for file in files:
    if file[-4:] == '.csv':
        data_files.append(file)
        
data_files
```

```python
data_dfs = []
for data_file in data_files:
    data_dfs.append(
        pd.read_csv(directory + data_file)
    )
    print(data_file)
    
for df in data_dfs:
    print(df.columns, '\n\n', df.size)
```

```python
def trim(string):
    return string.strip()
for index, row in data_dfs[0].iterrows():
    row['Location'] = row['Location'].strip()
    
data_dfs[0].Location = data_dfs[0].Location.map(trim)
data_dfs[0]
```

```python
consumption_countries = data_dfs[0].Location.unique()
consumption_countries
```

```python
density_countries = data_dfs[2]['Country Name'].unique()
density_countries
```

```python

def Diff(li1, li2): 
    '''
    # Python code t get difference of two lists 
    # Using set() 
    '''
    return (list(set(li1).symmetric_difference(set(li2))))
```

```python
delta = Diff(consumption_countries, density_countries)
delta
```
