```python
import pandas as pd
```

```python
annual_data = pd.read_csv("../Resources/annual_csv.csv")
annual_data.head()
```

```python
#conditionals to narrow down years
cleaned_annual_data = annual_data.loc[(annual_data["Year"] <= 2016) & 
                                (annual_data["Year"] >= 1980), :]
cleaned_annual_data
```

```python
#exporting clean table to a new csv for future access
cleaned_annual_data.to_csv('Temperature_Delta_1986 to 2016.csv')
```
